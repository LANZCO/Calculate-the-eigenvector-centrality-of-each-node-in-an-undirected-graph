```python
# importing librays
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# create graph
pro1_directed_graph = nx.DiGraph()
pro1_directed_graph.add_edges_from([('I', 'A'), ('I', 'B'), 
                                    ('I', 'C'), ('I', 'D'),
                                    ('I', 'E'), ('I', 'F'), 
                                    ('I', 'G'), ('I', 'H'), 
                                    ('A', 'I'), ('B', 'I'), 
                                    ('C', 'I'), ('D', 'I'), 
                                    ('E', 'I'), ('F', 'I'), 
                                    ('G', 'I'), ('H', 'I'),
                                    ('A','J')])

# make the graph
plt.figure(figsize=(8,8))
pos = nx.circular_layout(pro1_directed_graph)
nx.draw(pro1_directed_graph,pos,with_labels=True,
        node_size=900, alpha=0.9, arrows=True, font_size=10)
plt.title('Q1 Project Directed Graph')
plt.savefig('1.1.png')
plt.show()
```
